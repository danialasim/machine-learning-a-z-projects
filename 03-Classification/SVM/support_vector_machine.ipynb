# Support Vector Machine (SVM)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the SVM model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Visualising the Training set results

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

X_set = sc.inverse_transform(X_train)
y_set = y_train
colors = ['#FA8072', '#1E90FF']
cmap = ListedColormap(colors)

# Use larger step sizes for performance boost
x_min, x_max = X_set[:, 0].min() - 10, X_set[:, 0].max() + 10
y_min, y_max = X_set[:, 1].min() - 1000, X_set[:, 1].max() + 1000
X1, X2 = np.meshgrid(np.arange(x_min, x_max, 2),   # Changed step from 0.25 to 2
                     np.arange(y_min, y_max, 50))  # Changed step from 0.25 to 50

grid_points = np.c_[X1.ravel(), X2.ravel()]
Z = classifier.predict(sc.transform(grid_points)).reshape(X1.shape)

plt.figure(figsize=(10, 6))
plt.contourf(X1, X2, Z, alpha=0.75, cmap=cmap)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, label in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == label, 0], X_set[y_set == label, 1],
                c=colors[i], label=label, edgecolor='black')

plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualising the Test set results

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

X_set = sc.inverse_transform(X_test)
y_set = y_test
colors = ['#FA8072', '#1E90FF']
cmap = ListedColormap(colors)

# Use larger step size for faster mesh generation
x_min, x_max = X_set[:, 0].min() - 2, X_set[:, 0].max() + 2
y_min, y_max = X_set[:, 1].min() - 1000, X_set[:, 1].max() + 1000
X1, X2 = np.meshgrid(np.arange(x_min, x_max, 2),   # Changed from 0.25 to 2
                     np.arange(y_min, y_max, 50))  # Changed from 0.25 to 50

grid_points = np.c_[X1.ravel(), X2.ravel()]
Z = classifier.predict(sc.transform(grid_points)).reshape(X1.shape)

plt.figure(figsize=(10, 6))
plt.contourf(X1, X2, Z, alpha=0.75, cmap=cmap)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, label in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == label, 0], X_set[y_set == label, 1],
                color=colors[i], label=label, edgecolor='black')

plt.title('SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
